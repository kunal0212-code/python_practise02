```python
# What it does: Adds one item to the very end of the list.

# Example:

# Start with a list of groceries
groceries = ['milk', 'eggs']
print(f"Before: {groceries}")

# You remember you also need bread
groceries.append('bread')

print(f"After: {groceries}")
```

```python
# What it does: Adds one item at a specific position (index) you choose. Remember, Python counting starts at 0!

# Example:

# Python

# Start with a list
guests = ['Alice', 'Charlie']
print(f"Before: {guests}")

# You want to add 'Bob' at index 1 (between Alice and Charlie)
guests.insert(1, 'Bob')

print(f"After: {guests}")
```

```python
# What it does: Adds all the items from another list (or any iterable) to the end of your current list.

# Example:

# Python

# Your fruits for the week
fruits = ['apple', 'banana']
print(f"Before: {fruits}")

# You buy more fruits
more_fruits = ['orange', 'grape', 'mango']
fruits.extend(more_fruits)

print(f"After: {fruits}")
```

```python
# What it does: Removes the first item in the list that matches the value you provide.

# Example:

# Python

# A list with a duplicate
letters = ['a', 'b', 'c', 'b', 'd']
print(f"Before: {letters}")

# Remove the *first* 'b' you find
letters.remove('b')

print(f"After: {letters}")
```

```python
# What it does: Removes the item at a specific position (index). A special feature is that it also returns the item it removed, so you can save it.

# If you don't provide an index, it removes the last item.

# Example:

# Python

# A list of tasks
tasks = ['wash car', 'buy groceries', 'call mom']
print(f"Before: {tasks}")

# Remove and save the task at index 1
done_task = tasks.pop(1)

print(f"After: {tasks}")
print(f"Completed task: {done_task}")
```

```python
# What it does: Removes all items from the list, leaving it completely empty.

# Example:

# Python

# A list full of numbers
numbers = [1, 2, 3, 4, 5]
print(f"Before: {numbers}")

# Empty the list
numbers.clear()

print(f"After: {numbers}")
```

```python
# What it does: Tells you the position (index) of the first item that matches the value you're looking for.

# Note: It will cause an error if the item is not in the list.

# Example:

# Python

# A list of players
players = ['Anna', 'Ben', 'Chris', 'Dana']

# Find the position of 'Chris'
position = players.index('Chris')

print(f"'Chris' is at index: {position}")
```

```python
# What it does: Counts how many times a specific value appears in the list.

# Example:

# Python

# A list of votes
votes = ['yes', 'no', 'yes', 'yes', 'maybe', 'no']

# Count the number of 'yes' votes
yes_votes = votes.count('yes')

print(f"Total 'yes' votes: {yes_votes}")
```

```python
# What it does: Sorts the items in the list permanently (this is called "in-place").

# For numbers: sorts from smallest to largest.

# For strings: sorts alphabetically.

# Example:

# Python

# A list of numbers
scores = [88, 95, 72, 100, 81]
print(f"Before: {scores}")

scores.sort() # Sorts the list in-place

print(f"After: {scores}")
```

```python
# What it does: Reverses the order of the items in the list permanently ("in-place").

# Example:

# Python

# A countdown
countdown = [1, 2, 3, 4, 5]
print(f"Before: {countdown}")

countdown.reverse() # Reverses the list in-place

print(f"After: {countdown}")
```

```python
# What it does: Creates a shallow copy (a new, independent duplicate) of your list.

# Why it's important: If you just use =, you're not making a new list. Both variables will point to the same list. copy() creates a truly separate list.

# Example:

# Python

# Original list
list_a = ['x', 'y', 'z']

# Create a *real* copy
list_b = list_a.copy()

# Now, add something to the new list
list_b.append('w')

print(f"List B (the copy): {list_b}")
print(f"List A (original): {list_a}")
```
